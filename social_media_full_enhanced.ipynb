## 2. Enhanced Implementation (Advanced Features)

```python
from datetime import datetime

class Post:
    def __init__(self, author, content):
        if not content.strip():
            raise ValueError("Post content cannot be empty.")
        self.author = author
        self.content = content
        self.likes = 0
        self.comments = []
        self.timestamp = datetime.now()
        self.like_history = set()

    def like(self, user=None):
        self.likes += 1
        if user:
            self.like_history.add(user)

    def unlike(self, user=None):
        if self.likes > 0:
            self.likes -= 1
            if user and user in self.like_history:
                self.like_history.remove(user)

    def add_comment(self, comment, user=None):
        if comment.strip():
            if user:
                self.comments.append(f"{user}: {comment}")
            else:
                self.comments.append(comment)

    def edit_content(self, new_content):
        if new_content.strip():
            self.content = new_content

    def display(self):
        print(f"Author: {self.author} | Time: {self.timestamp.strftime('%Y-%m-%d %H:%M:%S')}")
        print(f"Content: {self.content}")
        print(f"Likes: {self.likes}")
        if self.like_history:
            print(f"Liked by: {', '.join(self.like_history)}")
        print("Comments:")
        for c in self.comments:
            print(f"- {c}")
        print("-" * 40)
```

```python
class SocialMediaApp:
    def __init__(self):
        self.posts = []

    def create_post(self, author, content):
        try:
            self.posts.append(Post(author, content))
        except ValueError as e:
            print(f"Error: {e}")

    def view_feed(self):
        if not self.posts:
            print("No posts yet.")
        for i, post in enumerate(self.posts):
            print(f"[Post {i}]")
            post.display()

    def like_post(self, index, user=None):
        if 0 <= index < len(self.posts):
            self.posts[index].like(user)
        else:
            print("Invalid post index.")

    def unlike_post(self, index, user=None):
        if 0 <= index < len(self.posts):
            self.posts[index].unlike(user)
        else:
            print("Invalid post index.")

    def comment_on_post(self, index, comment, user=None):
        if 0 <= index < len(self.posts):
            self.posts[index].add_comment(comment, user)
        else:
            print("Invalid post index.")

    def delete_post(self, index):
        if 0 <= index < len(self.posts):
            removed = self.posts.pop(index)
            print(f"Deleted post by {removed.author}")
        else:
            print("Invalid post index.")

    def search_by_author(self, author):
        found = [post for post in self.posts if post.author == author]
        if not found:
            print(f"No posts found for author '{author}'")
        for post in found:
            post.display()
```

### Example Usage

```python
app = SocialMediaApp()
app.create_post("Alice", "Hello world!")
app.create_post("Bob", "Learning Python is fun!")
app.like_post(0, user="Bob")
app.comment_on_post(0, "Nice post!", user="Bob")
app.view_feed()
app.posts[0].edit_content("Hello world! Edited.")
app.delete_post(1)
app.search_by_author("Alice")
app.unlike_post(0, user="Bob")
app.view_feed()
```
